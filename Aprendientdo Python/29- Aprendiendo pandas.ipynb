# ANALISIS DE CORRELACION ENTRE PASES Y GOLES


```python
# Para quitar el signo "%" de la columna posesion para luego convertirlo en un int, debo realizar un mapeo en la columna y 
#aplicar el metodo "replace()":

posesion = ds2['Posesión E1'].map(lambda x: x.replace('%','')).astype(int)
goles = ds2['Goles E1']



```

```python
import numpy as np

np.corrcoef(posesion, goles)
```

```python
posesion2 = ds2['Posesión E2'].map(lambda x: x.replace('%','')).astype(int)
goles2 = ds2['Goles E2']

np.corrcoef(posesion2, goles2)
```
